평가 방법: RMSLE이 대회의 순위는 **RMSLE (Root Mean Squared Logarithmic Error)**로 결정됩니다.$$\sqrt{\frac{1}{n} \sum_{i=1}^n (\log(p_i + 1) - \log(a_i + 1))^2}$$특징 1: 예측값과 실제값의 '차이'보다 **'비율'**에 집중합니다.특징 2: 실제값보다 낮게 예측했을 때(Underestimation) 더 큰 페널티를 부여합니다.특징 3: 로그를 취하기 때문에 아웃라이어(갑자기 대여량이 튀는 경우)의 영향력을 줄여줍니다.

대회 규칙 및 주의사항`
과거 데이터만 사용: 예측 시점보다 미래의 데이터를 사용해서는 안 됩니다.

외부 데이터 주의: 이 데이터는 이미 공개된 과거 데이터이므로, 인터넷에서 정답을 찾아 제출하는 '치팅'을 하지 말라는 경고가 있습니다.
